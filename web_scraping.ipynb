```python
def flatten(l):
    l_prime=[]
    for i in l:
        for j in i:
            l_prime.append(j)
    return l_prime
```

```python

def text_extractor(web_address):   
    # Send the request to the Wikipedia page
    response = requests.get(
    url=web_address,
    )

    # Parse the response content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the main content of the Wikipedia article
    content_div = soup.find("div", class_="mw-parser-output")

    # Extract all the text from paragraphs inside this div
    paragraphs = content_div.find_all('p')

    # Combine all paragraphs into a single text
    article_text = "\n".join([para.get_text() for para in paragraphs])

    # Print the extracted article text
    return article_text

    
```

```python
def extract_urls_webpage(web_address):
    # Send the request to the Wikipedia page
    response = requests.get(
    url="https://en.wikipedia.org/wiki/Category:Lists_of_cars",
    )

    # Parse the response content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all hyperlink elements (anchor tags)
    links = soup.find_all('a')

    # Extract the href attribute from each hyperlink
    urls = [link.get('href') for link in links if link.get('href')]
    return urls
    
```

```python
def extract_urls_webpage_new(web_address, exclude_urls):
    # Send the request to the specified web page
    response = requests.get(web_address)

    # Parse the response content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Base URL for converting relative URLs to absolute
    base_url = "https://en.wikipedia.org"

    # Find all hyperlink elements (anchor tags)
    links = soup.find_all('a')

    # Extract the href attribute from each hyperlink and convert to absolute URLs
    urls = [urljoin(base_url, link.get('href')) for link in links if link.get('href')]

    # Filter out URLs that are in the exclude_urls list
    filtered_urls = [url for url in urls if url not in exclude_urls]

    return filtered_urls
    
```

```python
urls=extract_urls_webpage("https://en.wikipedia.org/wiki/Category:Lists_of_cars")
print('1 is done')
urls_1st=urls[53:159]
# Print all the URLs
urls_2nd=[]
for url in tqdm.tqdm(urls[53:159]):
    url_2nd=extract_urls_webpage_new("https://en.wikipedia.org"+url,set(urls_2nd+urls_1st))
    urls_2nd.extend(url_2nd)

print('2 is done')
urls_3rd=[]
for url in tqdm.tqdm(urls_2nd):
    url_3rd=extract_urls_webpage_new("https://en.wikipedia.org"+url,set(urls_2nd+urls_1st+urls_3rd))
    urls_3rd.append(url_3rd)

print('3 is done')

urls_4th=[]
for url in tqdm.tqdm(urls_3rd):
    url_4th=extract_urls_webpage_new("https://en.wikipedia.org"+url,set(urls_2nd+urls_1st+urls_3rd+urls_4th))
    urls_4th.append(url_4th)

print('4 is done')


urls_5th=[]
for url in tqdm.tqdm(urls_4th):
    url_5th=extract_urls_webpage_new("https://en.wikipedia.org"+url,set(urls_2nd+urls_1st+urls_3rd+urls_4th+urls_5th))
    urls_5th.append(url_5th)

print('5 is done')
combined_urls=set(urls_1st+urls_2nd+urls_3rd+urls_4th+urls_5th)


n_chunk=len(combined_urls)//10000
n_1=len(combined_urls)%10000
for i in range(n_chunk):
    text_s=[]
    chunk_urls=combined_urls[i*10000:(i+1)*10000]
    for url in chunk_urls:
        text_s.append(text_extractor(url))
    pd.DataFrame({'url':chunk_urls,'text':text_s}).to_csv('car_data_'+str(i)+'.csv')

text_s=[]
chunk_urls=combined_urls[n_chunk*10000:]
for url in chunk_urls:
    text_s.append(text_extractor(url))
pd.DataFrame({'url':chunk_urls,'text':text_s}).to_csv('car_data_'+str(n_chunk)+'.csv')
```
